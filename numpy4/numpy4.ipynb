## Image Manipulation

```python
import matplotlib.pyplot as plt
img = plt.imread('dog.jpeg')
plt.imshow(img)
```

```python
img.shape
```

```python
img_b = img.copy()
img_b[:,:,(0,1)] = 0
plt.imshow(img_b)
```

```python
img_b = img.copy()
plt.imshow(img_b[:,:,0], cmap = 'gray')
```

```python
img_b = img.copy()
plt.imshow(np.rot90(img_b))
```

```python
img_b = img.copy()
plt.imshow(img_b[::-1,:,:])
```

```python
plt.imshow(img[::-1,::-1,:])
```

```python
plt.imshow(img)
```

```python
plt.imshow(img[0:250,200:400,:])
```

```python
plt.imshow(np.tile(img, (3,3,1)))
```

```python
img_masked = np.where(img>150, 250, 0)
plt.imshow(img_masked)
```

```python
plt.imshow(img-100)
```

```python
plt.imshow(255-img)
```
